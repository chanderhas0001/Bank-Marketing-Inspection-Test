# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv("C:/Users/Dell/Desktop/SKILLFIED MENTOR/first/downloaded-file/bankmarketing.csv")
#print(df)
# Display the first few rows
df.head()
```

```python

# Check for missing values and data types
# Check data types of each column
print("Data Types:")
print(df.dtypes)

# Check for missing values
print("Missing Values:")
print(df.isnull().sum())


```

```python

# Summary statistics
stats_summary=df.describe()
print(stats_summary)

```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable

target_column = 'age'  
sns.histplot(df[target_column])

# Labeling
plt.title(f'Distribution of {target_column}')
plt.xlabel('target_colum')
plt.ylabel('Frequency')

# Show plot
plt.show()

```

```python

# Correlation matrix for numerical features
numeric_df = df.select_dtypes(include=['number'])
corr_matrix=numeric_df.corr()
print(corr_matrix)

```
